```python
import os
import shutil

# Define the source and destination directories
source_dir = "/content/drive/MyDrive/knowledge/University/Master/Thesis/denoised/"
dest_base_dir = "/content/drive/MyDrive/knowledge/University/Master/Thesis/denoised_ordered/"

# Create the destination directory if it doesn't exist
if not os.path.exists(dest_base_dir):
    os.makedirs(dest_base_dir)

# Create the 5 subdirectories
subdirs = ["Static-x20", "Static-x40", "1.4Pa-x20", "1.4Pa-x40", "flow3-x20"]
for subdir in subdirs:
    subdir_path = os.path.join(dest_base_dir, subdir)
    if not os.path.exists(subdir_path):
        os.makedirs(subdir_path)

# Function to determine the destination folder for a file
def get_dest_folder(filepath, filename):
    # Extract the parent directory name
    parent_dir = os.path.basename(os.path.dirname(filepath))

    # Skip files from the flow3_1.4Pa_18h directory
    if "flow3_1.4Pa_18h" in filepath:
        return None

    # Only process files with "Flat" in their name
    if "Flat" not in filename:
        return None

    # Check for magnification in the filename
    is_40x = "40x" in filename
    is_20x = "20x" in filename or (not is_40x)  # Default to 20x if 40x not found

    # Determine the category
    if "Static-A-1" in parent_dir or "Static-A-2" in parent_dir or "Static" in filename or "0Pa" in filename:
        return "Static-x40" if is_40x else "Static-x20"

    elif "1.4Pa-A-1" in parent_dir or "1.4Pa-A-2" in parent_dir or "1.4Pa" in filename:
        return "1.4Pa-x40" if is_40x else "1.4Pa-x20"

    elif "flow3" in parent_dir or "flow3" in filename:
        return "flow3-x20"  # All flow3 files go to flow3-x20

    # If no match, return None
    return None

# Track statistics
files_copied = 0
files_skipped = 0

# Iterate through files in the source directory and its subdirectories
for root, dirs, files in os.walk(source_dir):
    for filename in files:
        # Skip hidden files
        if filename.startswith('.'):
            continue

        # Get the full path of the source file
        source_file_path = os.path.join(root, filename)

        # Determine the destination folder
        dest_folder = get_dest_folder(source_file_path, filename)

        # If a destination folder was found, copy the file
        if dest_folder:
            dest_file_path = os.path.join(dest_base_dir, dest_folder, filename)
            # Copy the file with metadata
            shutil.copy2(source_file_path, dest_file_path)
            print(f"Copied {filename} to {dest_folder}")
            files_copied += 1
        else:
            reason = ""
            if "flow3_1.4Pa_18h" in source_file_path:
                reason = "(from flow3_1.4Pa_18h directory)"
            elif "Flat" not in filename:
                reason = "(doesn't contain 'Flat')"
            print(f"Skipped {filename} {reason}")
            files_skipped += 1

print(f"File organization complete! Copied {files_copied} files, skipped {files_skipped} files.")
```

```python
import os
import shutil

# Define source and destination paths
source_dir = "/content/drive/MyDrive/knowledge/University/Master/Thesis/denoised/flow3_1.4Pa_18h"
dest_dir = "/content/drive/MyDrive/knowledge/University/Master/Thesis/denoised_ordered/flow3-x20"

# Create destination directory if it doesn't exist
if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)
    print(f"Created destination directory: {dest_dir}")

# Get list of files in source directory containing "Flat"
flat_files = [f for f in os.listdir(source_dir) if "Flat" in f]

if not flat_files:
    print("No files with 'Flat' in their name found.")
else:
    # Copy each file
    copied_count = 0
    for file in flat_files:
        source_path = os.path.join(source_dir, file)
        dest_path = os.path.join(dest_dir, file)

        shutil.copy2(source_path, dest_path)
        print(f"Copied: {file}")
        copied_count += 1

    print(f"Successfully copied {copied_count} files to {dest_dir}")
```
